```python
import math


class Creature:
    def __init__(self, birth_rate, death_rate, mutation_factor, population_init):
        self.birth_rate = birth_rate
        self.death_rate = death_rate
        self.mutation_factor = mutation_factor
        self.population_init = population_init

    # P(t) = e^{(B-D)t}(P_0 + \frac{m}{(B-D)^2}) - \frac{m}{(B-D)^2} - \frac {mt}{B-D}
    def calculate_population(self, t):
        k = self.birth_rate - self.death_rate
        p_0 = self.population_init
        m = self.mutation_factor
        p = math.exp(k * t) * (p_0 + (m / (k ** 2)) - (m / (k ** 2)) - (m * t / k))
        return p
```

```python
import matplotlib.pyplot as plt
from matplotlib import animation
import numpy as np
import seaborn as sns
```

```python
deer_mutated = Creature(0.45, 0.444, 0.01, 100)
deer = Creature(0.45, 0.444, 0, 100)
x, y = [], []
y_m = []
# simulate out for 100 steps
for n in range(100):
    x.append(n)
    y.append(deer.calculate_population(n))
    y_m.append(deer_mutated.calculate_population(n))
```

```python

fig = plt.figure(figsize=(10, 6))
plt.xlim(0, 100)
plt.ylim(0,200)
plt.xlabel('Years', fontsize=20)
plt.ylabel('Deer Population', fontsize=20)
plt.plot(x,y_m)
plt.title('Deer population over 100 years', fontsize=20)
```
